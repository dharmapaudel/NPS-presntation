## Simple plotting 

# Ploting of a function 

```python
from pylab import plot, show, title, xlabel, ylabel
from numpy import linspace, sin 
x = linspace(0,10,100)
y = sin(x)
plot(x,y) 
title("Sine function")
xlabel("x")
ylabel("sin(x)")
show() 
```

# Plotting of a given data 

```python
from numpy import loadtxt
from pylab import plot, show
data = loadtxt("testdata",float)
x = data[:,0]
y = data[:,1]
plot(x,y)
show()
```

#  Plottting solar irradiation data at a given place 28.585 N, 81.637  E 
# (Surkhet  airport) from 2005 - 2016 (AD) 


```python
import pandas as pd
import numpy as np
import random as random
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
%matplotlib inline
sns.set()
```

```python
df2 = pd.read_csv('month.csv', names = ['Year','month','Rad'])
                  
```

```python
df2.head(13)
```

```python
df2["Year"] = pd.to_datetime(df2["Year"])
df2.info()
```

```python
plt.figure(figsize=[12,6])
plt.plot(df2.Year, df2.Rad)

plt.title("Monthly solar irradiation estimates ")
plt.xlabel("Year")
plt.ylabel("Monthly irradiation level in kWh/$m^2$")
plt.show()
```
